```python
def to_datetime(x):
    """
    Mappting string date to datetime class
    When date is missing or wrong return None

    Args:
        x: string date
    Returns:
        x: datetime
    """
    try:
        x = list(map(int, x.replace('|', ' ').split()))
        return datetime.datetime(*x)
    except:
        return 0
```

```python
date_data['MI symptom recent time'] = date_data['MI symptom recent time'].apply(to_datetime)
date_data['This hospital Arrival time'] = date_data['This hospital Arrival time'].apply(to_datetime)
```
